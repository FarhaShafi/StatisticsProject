## 4. Identify is there a significant difference in spending between male and female customers

```python
# Average spending by gender
avg_purchase_gender = data.groupby('Gender')['PurchaseAmount'].mean()
print(f"Average spending by gender :\n{avg_purchase_gender}")
```

```python
# Perform statistical test
male_spending = data[data['Gender'] == 'Male']['PurchaseAmount']
female_spending = data[data['Gender'] == 'Female']['PurchaseAmount']
# Independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(male_spending, female_spending)
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in spending between male and female customers")
else:
    print("There is no significant difference in spending between male and female customers")
```

## 5. Identify is there a relationship between ProductCategory and customer churn

```python
# Create contingency table
contingency_table = pd.crosstab(data['ProductCategory'], data['Churn'])
print("Contingency Table:\n", contingency_table)
```

```python
from scipy.stats import chi2_contingency

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-square value:", chi2)
print("p-value:", p)
print(dof)
print(expected)
```

```python
alpha = 0.05
if p < alpha:
    print("Product Category is significantly related to Churn")
else:
    print("No significant relationship found")
```

## 6. Does PurchaseAmount vary significantly across different regions

```python
region_list = data['Region'].unique().tolist()
print(region_list)
```

```python
sns.boxplot(x = 'Region', y = 'PurchaseAmount', data = data)
```

```python
# One-way ANOVA test
groups = [data[data['Region'] == reg]['PurchaseAmount'] for reg in region_list]
f_statistics, p_value = stats.f_oneway(*groups)
print("F-statistic:", f_statistics)
print("P-value:", p_value)
```

```python
# Interpretation
alpha = 0.5
if p_value <= alpha:
    print("PurchaseAmount vary significantly across different regions")
else:
    print("PurchaseAmount does not vary significantly across different regions")
```

## 7. Identify which email campaign (A or B) performed better in terms of average PurchaseAmount

```python
data['CampaignGroup'].unique()
```

```python
campaign_a = data[data['CampaignGroup'] == 'A']['PurchaseAmount']
campaign_b = data[data['CampaignGroup'] == 'B']['PurchaseAmount']

# Calculate the average purchase amount
mean_a = np.mean(campaign_a)
mean_b = np.mean(campaign_b)
print(f"Average of campaign A = {mean_a}")
print(f"Average of campaign B = {mean_b}")
```

```python
# Perform the independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(campaign_a, campaign_b)
print(f"T-Statistic : {t_statistic}")
print(f"P-Value : {p_value}")
```

```python
alpha = 0.5
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    if mean_a > mean_b:
        print("Conclusion: Campaign A performed significantly better than Campaign B in terms of average purchase amount.")
    else:
        print("Conclusion: Campaign B performed significantly better than Campaign A in terms of average purchase amount.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: There is no significant difference in the average purchase amounts between Campaign A and Campaign B.")
```

## 9. What insights can we gain by applying the Central Limit Theorem?

```python
means = []
samples = 1000
size = 30

for _ in range(samples):
    # Draw a random sample with replacement
    sample = np.random.choice(data['PurchaseAmount'], size=size, replace=True)
    means.append(np.mean(sample))

# Plotting the results to show it's normally distributed
plt.hist(means, bins=30, density=True, alpha=0.7)
plt.title('Distribution of Means with Central Limit Theorem')
plt.xlabel('Average Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

## 10. What is the 95% confidence interval for the average PurchaseAmount?

```python
mean_of_means = np.mean(means)
std_error = np.std(means) / np.sqrt(size)
conf_interval = (mean_of_means - 1.96 * std_error, mean_of_means + 1.96 * std_error)
print(f"95% Confidence Interval for average PurchaseAmount: {conf_interval}")
```
